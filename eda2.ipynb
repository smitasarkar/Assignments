```python
# Count the number of instances in each class
class_counts = df['target'].value_counts()

# Print the class counts
print(class_counts)
```

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE

# Split the dataframe into features and target variables
X = df.drop('target', axis=1)
y = df['target']

# Over-sample the minority class
X_resampled, y_resampled = SMOTE().fit_resample(X, y)

# Combine the over-sampled data with the original data
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
# Count the number of instances in each class
class_counts = df_resampled['target'].value_counts()

# Print the class counts
print(class_counts)
```

```python
df_resampled.to_csv('smote_data.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import ADASYN

# Split the dataframe into features and target variables
X = df.drop('target', axis=1)
y = df['target']

# Create an instance of the ADASYN class
adasyn = ADASYN()

# Over-sample the minority class
X_resampled, y_resampled = adasyn.fit_resample(X, y)

# Combine the over-sampled data with the original data
df_adasyn = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
# Count the number of instances in each class
class_counts = df_adasyn['target'].value_counts()

# Print the class counts
print(class_counts)
```

```python
df_adasyn.to_csv('adasyn_data.csv', index=False)
```
